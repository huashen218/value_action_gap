```python
import json
import pandas as pd

# Read the JSONL correctly to make each JSON object a row
results = []
with open("../../outputs/task2_prompt_evaluation_gpt_4o_mini.jsonl", "r") as f:
    for line in f:
        # Each line is a dictionary that should be a row
        results.append(pd.json_normalize(json.loads(line)))

# Concatenate all the individual row dataframes
df = pd.concat(results, ignore_index=True)

df.head()
```

```python
for index, row in df.iterrows():
    res = [0,0]
    for i in range(8):
        if row[f"evaluation_{i}.action"] == "Option 1":
            res[0] += 1
        else:
            res[1] += 1
    df.at[index, "majority_vote"] = "Option 1" if res[0] > res[1] else "Option 2" if res[1] > res[0] else "Neither"
    df.at[index, "majority_vote_count"] = int(max(res))

df.head()
```

```python
# percent agreement
df["percent_agreement"] = df["majority_vote_count"] / 8


# average percent agreement
df["percent_agreement"].mean()
```

```python
# evaulation_{i}.aciton that most agrees with majority vote
for i in range(8):
    df[f"evaluation_{i}.agreement"] = df[f"evaluation_{i}.action"] == df["majority_vote"]
    print(f"evaluation_{i}.agreement: {df[f'evaluation_{i}.agreement'].mean()}")
```
